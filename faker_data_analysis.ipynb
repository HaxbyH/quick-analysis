# Simple Data Analysis with Faker

This notebook generates fake customer data using [Faker](https://faker.readthedocs.io/en/master/) and performs simple data analysis using external libraries such as Pandas, Numpy, Matplotlib, and Seaborn.

```python
from faker import Faker
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

fake = Faker()
```

## Generate Fake Data

We'll generate a dataset of fake customers with attributes like name, age, country, and income.

```python
def generate_fake_data(n=100):
    data = []
    for _ in range(n):
        profile = fake.simple_profile()
        data.append({
            'name': profile['name'],
            'sex': profile['sex'],
            'age': np.random.randint(18, 70),
            'country': fake.country(),
            'income': np.random.normal(50000, 15000)
        })
    return pd.DataFrame(data)

df = generate_fake_data(200)
df.head()
```

## Basic Statistics

```python
df.describe(include='all')
```

## Data Visualization

Let's visualize age and income distributions, and see if there's any relationship.

```python
plt.figure(figsize=(10,5))
sns.histplot(df['age'], bins=15, kde=True)
plt.title('Age Distribution')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(df['income'], bins=20, kde=True)
plt.title('Income Distribution')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='age', y='income', data=df, hue='sex')
plt.title('Age vs Income')
plt.show()
```
